次數分配表

```python
a = df.columns
pp2 = []
for j in a:
  try:
    print(df[j].groupby(df[j]).count())
    pp2.append(df[j].groupby(df[j]).count())
  except:
    print("KeyError")
with open('次數table_3.txt', 'w', encoding = 'utf-8') as f:
  f.write(str(pp2))
  f.close
```

Merge

```python
left = pd.DataFrame({"key": ["A","B"], "lval": [1,2]})
print(left)
right = pd.DataFrame({"key": ["A","B"], "rval": [3,4]})
print(right)
# according to the condition of key to merge
pd.merge(right, left, on="key")
```
